```python
#dropping due to low corellation with target varaiable
dataframe = dataframe.drop(['children', 'male', 'northeast', 'northwest', 'southeast','southwest'], axis = 1)
dataframe
```

```python
#converting expenses data to int
dataframe['expenses'] = dataframe['expenses'].astype('int')
dataframe
```

```python
from sklearn.model_selection import train_test_split
df_train,df_test = train_test_split(dataframe, train_size=0.85, test_size=0.15, random_state = 1)
```

```python
df_train
```

```python
df_test
```

```python
df_train.shape
```

```python
df_test.shape
```

```python
from sklearn.linear_model import LinearRegression
X_train = df_train[['age','bmi',"smoke"]]

Y_train = df_train['expenses']


# for linear regression
lr = LinearRegression()
lr_model = lr.fit(X_train, Y_train)

# for logistic regression
from sklearn.linear_model import LogisticRegression
lg = LogisticRegression()
lg_model = lg.fit(X_train, Y_train)

# for polynomial regression
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree=8)
X_poly = poly_reg.fit_transform(X_train)  # transform training data into polynomial features

poly = LinearRegression()  
poly_model = poly.fit(X_poly, Y_train)  #fits the linear regression model using the polynomial features
```

```python
data = df_test.iloc[50:51]
actual_expenses = data['expenses']
data = data.drop(['expenses'],axis = 1)
data
```

```python
#linear model
print("Predicted expenses",lr_model.predict(data))
print("Actual expenses",actual_expenses)
```

```python
#logistic
print("Predicted expenses",lg_model.predict(data))
print("Actual expenses",actual_expenses)
```

```python
#polynomial regression 
data1 = poly_reg.fit_transform(data) #transforms this new data into the same polynomial feature space that was used to train poly_model.
predicted_expenses = poly_model.predict(data1)
print("Actual expenses",actual_expenses)
print("predicted expenses",predicted_expenses)
```

```python
#train acc
print("Linear regression = ",lr_model.score(X_train,Y_train)*100)
print("Logistic regression = ",lg_model.score(X_train, Y_train)*100)
print("Polynomial regression = ",poly_model.score(poly_reg.fit_transform(X_train), Y_train)*100)
```

```python
#test acc
test_data = df_test
Y_test = test_data['expenses']
X_test = test_data.drop(['expenses'],axis = 1)

print("Linear regression = ",lr_model.score(X_test, Y_test)*100)
print("Logistic regression = ",lg_model.score(X_test, Y_test)*100)
print("Polynomial regression = ",poly_model.score(poly_reg.fit_transform(X_test), Y_test)*100)
```

```python
# Testing
data = df_test.iloc[50:51]
actual_expenses = data.pop('expenses')

# # # print(data)
predicted_expenses = poly_model.predict(poly_reg.fit_transform(data))
print("Actual expenses",actual_expenses)
print("predicted Salary",predicted_expenses)
data
```
